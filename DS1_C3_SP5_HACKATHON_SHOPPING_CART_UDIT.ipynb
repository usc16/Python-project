```python
mydict
print("%10s %10s %10s %10s"%("S.no","ITEM","QUANTITY","PRICE"))
print("%10.2f %10s %10.2f %10.2f"%(1,"biscuits",mydict["biscuits"][0],mydict["biscuits"][1]))
print("%10.2f %10s %10.2f %10.2f"%(2,"cereals",mydict["cereals"][0],mydict["cereals"][1]))
print("%10.2f %10s %10.2f %10.2f"%(3,"rice",mydict["rice"][0],mydict["rice"][1]))
print("\n")

avail_cart=["biscuits","cereals","rice"]
avail_quantity=[]
avail_price=[20.50,30,15]

bill_cart=[]
bill_quantity=[]

while True:
    a=int(input("Enter 1 for biscuits,2 for cereals,3 for rice ,4 to exit: "))
    if a==1:
        if mydict["biscuits"][0]>0:
            b=int(input("how many units of biscuits you want to purchase: "))
            mydict["biscuits"][0]-=b
#If the item is out of stock it will display the same            
        else:
            print("item is out of stock")
            bill_quantity.append(0)
#If item is already in bill cart we will first calculate index at which it is present            
#and then append the quantity corresponding to that index in bill quantity cart             
        if "biscuits" in bill_cart:
            print("This item is already present in cart")
            ix4=bill_cart.index("biscuits")
            bill_quantity[ix4]+=b
#If item is not there in the bill cart it will append the cart as well as quantity in the respective list            
        else:
            print("new selection")
            bill_cart.append("biscuits")
            bill_quantity.append(b)
    if a==2:
        if mydict["cereals"][0]>0:
            b=int(input("how many units of cereals you want to purchase: "))
            mydict["cereals"][0]-=b
        else:
            print("item is out of stock")
            bill_quantity.append(0)
        if "cereals" in bill_cart:
            print("Item is already in the cart you can add more")
            ix5=bill_cart.index("cereals")
            bill_quantity[ix5]+=b
        else:
            print("new selection")
            bill_cart.append("cereals")
            bill_quantity.append(b)
    if a==3:
        if mydict["rice"][0]>0:
            b=int(input("how much rice you want to purchase: "))
            mydict["rice"][0]-=b
        else:
            print("item is out of stock")
            bill_quantity.append(0)
        if "rice" in bill_cart:
            print("Item is already in the cart you can add more")
            ix6=bill_cart.index("rice")
            bill_quantity[ix6]+=b
        else:
            print("new selection")
            bill_cart.append("rice")
            bill_quantity.append(b)            
    if a==4:
        break
#For delivery details:
print("\n")
u_1=input("Enter your name: ")
v_1=input("Enter your address: ")
w_1=int(input("Enter distance from shop: "))
print("Delivery charges of 30 will be levied extra for more then 10 km of distance","\n")        
grand_total=0
#For shopping cart calculations

print("%10s %10s %10s %10s %10s"%("S.no","ITEM","QUANTITY","PRICE","TOTAL PRICE"))
for i in range(len(bill_cart)):
#Here for claculating the price of product in bill_cart we use for loop
#We match at what index our item in bill_cart is present in avail_cart and then take the price of same
    ix7=avail_cart.index(bill_cart[i])
    price=avail_price[ix7]
    total_price=price*bill_quantity[i]
    grand_total+=total_price
    print("%10.2f %10s %10.2f %10.2f %10.2f"%(i+1,bill_cart[i],bill_quantity[i],price,total_price))
print("\n")
if w_1>=10:
    print("Your total bill is",(grand_total+30))
else:
    print("Your total bill is",grand_total)
print("Name:",u_1)
print("Address:",v_1)
o=input("Do you want to continue shopping Y or NO: ")

    
    
```
